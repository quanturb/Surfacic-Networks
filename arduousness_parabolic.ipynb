# Arduousness calculation

Surfacic Networks

Barthelemy et al, PNAS Nexus, 2025

```python
import networkx as nx
import osmnx as ox
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

pd.set_option('display.max_rows', 100)

%matplotlib inline
ox.__version__
nx.__version__

# Initialize parameters
number_conf = 100
number_nodes = 100
rr = 1.0
avek = 10.0
r0 = rr * np.sqrt(avek / float(number_nodes))
print('r0= ', r0)

epsilon = 1.0 / 100.0
z_sup = 5.0
nz = 100
gamma_values = [0.1, 1.0, 3.0, 5.0]  # Values of gamma to plot

# Prepare figure for plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Loop over different gamma values
for gamma in gamma_values:
    print(f"Processing for gamma={gamma}")

    list_z = []
    list_ard = []
    
    for iz in range(0, nz):
        z_max = z_sup * float(iz) / float(nz)
        print('zmax= ', z_max)

        counter = 0
        list_ratl = []

        for conf in range(number_conf):
            counter += 1
            print(f'----- conf #{counter} zmax={z_max}')
            
            # Generate random points
            r = np.random.uniform(low=0, high=1, size=number_nodes)  # radius
            theta = np.random.uniform(low=0, high=2 * np.pi, size=number_nodes)  # angle

            x = np.sqrt(r) * np.cos(theta)
            y = np.sqrt(r) * np.sin(theta)
            z = z_max * (1.0 - np.square(x) - np.square(y))

            # Read positions
            nplist = np.arange(start=1, stop=number_nodes + 1, step=1)
            df_pos = pd.DataFrame({'node_i': nplist, 'x': x, 'y': y, 'z': z})

            # Create adjacency matrix
            df_adj = pd.DataFrame(columns=('node_i', 'node_j', 'dz', 'd3d', 'd2d', 'absdz', 'weight'))
            icount = 0
            for i in range(1, number_nodes + 1):
                for j in range(1, number_nodes + 1):
                    if i != j:
                        dx = x[j - 1] - x[i - 1]
                        dy = y[j - 1] - y[i - 1]
                        dz = z[j - 1] - z[i - 1]
                        d2d_val = np.sqrt(dx**2 + dy**2)
                        d3d_val = np.sqrt(dx**2 + dy**2 + dz**2)

                        if d3d_val < r0:  # Definition of RGG in 3D
                            icount += 1
                            absdz_val = abs(dz * np.heaviside(dz, 0.0))

                            if dz <= 0.0:
                                weight_val = d3d_val
                            else:
                                weight_val = gamma * dz + d3d_val

                            df_adj.loc[icount] = [i, j, dz, d3d_val, d2d_val, absdz_val, weight_val]

            df_adj = df_adj.astype({'node_i': 'int', 'node_j': 'int'})

            # Create weighted graph
            G3d = nx.from_pandas_edgelist(df_adj, 'node_i', 'node_j',
                                          create_using=nx.Graph(), edge_attr='d3d')

            Ggen = nx.from_pandas_edgelist(df_adj, 'node_i', 'node_j',
                                           create_using=nx.DiGraph(), edge_attr='weight')

            # Shortest paths analysis
            for source in range(1, number_nodes):
                for target in range(source + 1, number_nodes + 1):
                    if source in G3d.nodes and target in G3d.nodes and nx.has_path(G3d, source, target):
                        list_nodes3d = nx.dijkstra_path(G3d, source, target, weight='d3d')
                        list_nodesgen = nx.dijkstra_path(Ggen, source, target, weight='weight')

                        suml = sum(
                            np.sqrt((x[second - 1] - x[first - 1])**2 + (y[second - 1] - y[first - 1])**2 +
                                    (z[second - 1] - z[first - 1])**2) 
                            for first, second in zip(list_nodes3d, list_nodes3d[1:])
                        )

                        suml_lazy = sum(
                            np.sqrt((x[second - 1] - x[first - 1])**2 + (y[second - 1] - y[first - 1])**2 +
                                    (z[second - 1] - z[first - 1])**2)
                            for first, second in zip(list_nodesgen, list_nodesgen[1:])
                        )

                        # Lazyness analysis
                        ratio_l = suml_lazy / suml
                        list_ratl.append(ratio_l)

        # Compute arduousness
        arduousness = sum(list_ratl) / len(list_ratl) - 1.0
        list_z.append(z_max)
        list_ard.append(arduousness)
        print(f'arduousness= {z_max}, {arduousness}')

    # Plot the results for the current gamma
    df_ard = pd.DataFrame({'zmax': np.array(list_z), 'ard': np.array(list_ard)})
    sns.lineplot(x=df_ard['zmax'], y=df_ard['ard'], label=f'$\mu=${gamma}', ax=ax, alpha=0.7, linewidth=3)

# Customize and save plot
ax.set_xlabel(r'$z_{max}$', fontsize=24)
ax.set_ylabel(r'$A$', fontsize=24)
ax.tick_params(axis='both', labelsize=20)
ax.legend(loc='best', fontsize=20)

namefig1 = 'fig_arduous_multiple_gamma.pdf'
fig.savefig(namefig1)

plt.show()
```
